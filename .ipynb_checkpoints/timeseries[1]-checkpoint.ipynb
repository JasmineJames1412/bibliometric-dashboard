```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from statsmodels.tsa.arima.model import ARIMA
from prophet import Prophet
import warnings
warnings.filterwarnings("ignore")

# Load data
file_path = "author_details_output.xlsx"
citations_df = pd.read_excel(file_path, sheet_name='Citations Per Year')

# Clean data
citations_df = citations_df.dropna()
citations_df['Year'] = citations_df['Year'].astype(int)

# Split train (up to 2020) and test (2021â€“2025)
train = citations_df[citations_df['Year'] <= 2020]
test = citations_df[citations_df['Year'] > 2020]

# --- 1. Linear Regression ---
lr = LinearRegression()
lr.fit(train[['Year']], train['Citations'])
lr_preds = lr.predict(test[['Year']])

# --- 2. Holt Linear Trend ---
holt_model = ExponentialSmoothing(train['Citations'], trend='add', seasonal=None)
holt_fit = holt_model.fit()
holt_preds = holt_fit.forecast(len(test))

# --- 3. ARIMA ---
arima_model = ARIMA(train['Citations'], order=(2,1,2))
arima_fit = arima_model.fit()
arima_preds = arima_fit.forecast(len(test))

# --- 4. Random Forest Regressor with Lag Features ---
# Create lag features
def create_lag_features(df, lag=1):
    df_copy = df.copy()
    for i in range(1, lag+1):
        df_copy[f'lag_{i}'] = df_copy['Citations'].shift(i)
    return df_copy.dropna()

# Adding lag feature to the data
train_lag = create_lag_features(train, lag=3)  # Using 3 lags for example
test_lag = create_lag_features(test, lag=3)

# Train Random Forest on lagged features
rf = RandomForestRegressor(random_state=42)
rf.fit(train_lag[['Year', 'lag_1', 'lag_2', 'lag_3']], train_lag['Citations'])

# Predict on test data
rf_preds = rf.predict(test_lag[['Year', 'lag_1', 'lag_2', 'lag_3']])

# --- 5. Prophet ---
# Prepare data for Prophet
prophet_df = citations_df.rename(columns={'Year': 'ds', 'Citations': 'y'})
prophet_df['ds'] = pd.to_datetime(prophet_df['ds'], format='%Y')
train_prophet = prophet_df[prophet_df['ds'].dt.year <= 2020]
test_prophet = prophet_df[prophet_df['ds'].dt.year > 2020]

# Train Prophet model
prophet_model = Prophet(yearly_seasonality=False, daily_seasonality=False, weekly_seasonality=False)
prophet_model.fit(train_prophet)

# Forecast with Prophet for test years
future_test = prophet_model.make_future_dataframe(periods=len(test), freq='Y')
prophet_forecast = prophet_model.predict(future_test)
prophet_preds = prophet_forecast.iloc[-len(test):]['yhat'].values

# --- Evaluation ---
def evaluate(true, pred, model_name):
    mae = mean_absolute_error(true, pred)
    rmse = np.sqrt(mean_squared_error(true, pred))
    return {'Model': model_name, 'MAE': mae, 'RMSE': rmse}

results = [
    evaluate(test['Citations'], lr_preds, 'Linear Regression'),
    evaluate(test['Citations'], holt_preds, 'Holt Linear Trend'),
    evaluate(test['Citations'], arima_preds, 'ARIMA'),
    evaluate(test['Citations'], rf_preds, 'Random Forest'),
    evaluate(test['Citations'].values, prophet_preds, 'Prophet')
]

results_df = pd.DataFrame(results).sort_values('RMSE')
print("\nðŸ“Š Accuracy Comparison:")
print(results_df)

# --- Forecast 2026â€“2030 using best model ---
best_model = results_df.iloc[0]['Model']
future_years = np.arange(2026, 2031)

if best_model == 'ARIMA':
    final_model = ARIMA(citations_df['Citations'], order=(2,1,2)).fit()
    future_preds = final_model.forecast(len(future_years))

elif best_model == 'Holt Linear Trend':
    final_model = ExponentialSmoothing(citations_df['Citations'], trend='add', seasonal=None).fit()
    future_preds = final_model.forecast(len(future_years))

elif best_model == 'Linear Regression':
    final_model = LinearRegression()
    final_model.fit(citations_df[['Year']], citations_df['Citations'])
    future_preds = final_model.predict(future_years.reshape(-1, 1))

elif best_model == 'Random Forest':
    # Using Random Forest with lag features
    full_data_lag = create_lag_features(citations_df, lag=3)
    full_rf = RandomForestRegressor(random_state=42)
    full_rf.fit(full_data_lag[['Year', 'lag_1', 'lag_2', 'lag_3']], full_data_lag['Citations'])
    
    future_lag_features = np.column_stack([future_years, 
                                           np.zeros(len(future_years)),  # Placeholder for lag_1, lag_2, lag_3
                                           np.zeros(len(future_years)),
                                           np.zeros(len(future_years))])
    
    future_preds = full_rf.predict(future_lag_features)

elif best_model == 'Prophet':
    full_prophet_model = Prophet(yearly_seasonality=False, daily_seasonality=False, weekly_seasonality=False)
    full_prophet_model.fit(prophet_df)
    future_df = full_prophet_model.make_future_dataframe(periods=5, freq='Y')
    future_forecast = full_prophet_model.predict(future_df)
    future_preds = future_forecast.iloc[-5:]['yhat'].values

# --- Plot full forecast ---
plt.figure(figsize=(14, 7))
plt.plot(citations_df['Year'], citations_df['Citations'], label='Historical Citations', marker='o')
plt.plot(test['Year'], test['Citations'], label='Actual Test (2021â€“2025)', marker='o')
plt.scatter(future_years, future_preds, color='red', label=f'{best_model} Prediction (2026â€“2030)', zorder=5)

plt.title(f"Citations Forecast with {best_model} (2026â€“2030)")
plt.xlabel("Year")
plt.ylabel("Citations")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Print future predictions
forecast_table = pd.DataFrame({'Year': future_years, 'Predicted Citations': future_preds})
print(f"\nðŸ“ˆ {best_model} Forecast 2026â€“2030:")
print(forecast_table)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Load your dataset
file_path = 'author_details_output.xlsx'
pub_details_df = pd.read_excel(file_path, sheet_name='Publication Details')

# Clean and prepare data
pub_details_df_clean = pub_details_df.dropna(subset=['Publication Date'])
pub_details_df_clean['Publication Date'] = pub_details_df_clean['Publication Date'].astype(int)
pub_counts_by_year = pub_details_df_clean.groupby('Publication Date').size().reset_index(name='Publications')

# Use data from 1986 to 2025
pub_counts_86_25 = pub_counts_by_year[(pub_counts_by_year['Publication Date'] >= 1986) & 
                                      (pub_counts_by_year['Publication Date'] <= 2025)]

# Fit a linear regression model
X_train = pub_counts_86_25[['Publication Date']].values
y_train = pub_counts_86_25['Publications'].values

model = LinearRegression()
model.fit(X_train, y_train)

# Predict publication counts for all years from 1986 to 2030 (for trend line)
years_full = np.arange(1986, 2031).reshape(-1, 1)
predicted_trend = model.predict(years_full)

# Predict publication counts specifically for 2026 to 2030
future_years = np.array([[2026], [2027], [2028], [2029], [2030]])
future_predictions = model.predict(future_years)

# Plot historical data
plt.figure(figsize=(14, 7))
plt.plot(pub_counts_86_25['Publication Date'], pub_counts_86_25['Publications'], marker='o', color='blue', label='Historical Publications (1986-2025)')

# Plot trend line
plt.plot(years_full.flatten(), predicted_trend, color='skyblue', label='Forecast Trend')

# Plot predicted points for 2026-2030
plt.scatter(future_years.flatten(), future_predictions, color='red', label='Predicted 2026-2030', zorder=5)

# Title and labels
plt.title('Publication Forecast (2026-2030)', fontsize=16)
plt.xlabel('Year')
plt.ylabel('Number of Publications')
plt.grid(True)
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()

# Print predicted values
for year, pred in zip(future_years.flatten(), future_predictions):
    print(f"{year}: {pred:.0f} publications predicted")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from statsmodels.tsa.arima.model import ARIMA
from prophet import Prophet
import warnings
warnings.filterwarnings("ignore")

# Load data
file_path = "author_details_output.xlsx"
citations_df = pd.read_excel(file_path, sheet_name='Citations Per Year')

# Clean data
citations_df = citations_df.dropna()
citations_df['Year'] = citations_df['Year'].astype(int)

# Split train (up to 2020) and test (2021â€“2025)
train = citations_df[citations_df['Year'] <= 2020]
test = citations_df[citations_df['Year'] > 2020]

# --- 1. Linear Regression ---
lr = LinearRegression()
lr.fit(train[['Year']], train['Citations'])
lr_preds = lr.predict(test[['Year']])

# --- 2. Holt Linear Trend ---
holt_model = ExponentialSmoothing(train['Citations'], trend='add', seasonal=None)
holt_fit = holt_model.fit()
holt_preds = holt_fit.forecast(len(test))

# --- 3. ARIMA ---
arima_model = ARIMA(train['Citations'], order=(2,1,2))
arima_fit = arima_model.fit()
arima_preds = arima_fit.forecast(len(test))

# --- 4. Random Forest Regressor ---
rf = RandomForestRegressor(random_state=42)
rf.fit(train[['Year']], train['Citations'])
rf_preds = rf.predict(test[['Year']])

# --- 5. Prophet ---
prophet_df = train[['Year', 'Citations']].rename(columns={'Year': 'ds', 'Citations': 'y'})
prophet_df['ds'] = pd.to_datetime(prophet_df['ds'], format='%Y')

prophet = Prophet()
prophet.fit(prophet_df)

future_test = pd.DataFrame({'ds': pd.to_datetime(test['Year'], format='%Y')})
forecast_test = prophet.predict(future_test)
prophet_preds = forecast_test['yhat'].values

# --- Evaluation Function ---
def evaluate(true, pred, model_name):
    mae = mean_absolute_error(true, pred)
    rmse = np.sqrt(mean_squared_error(true, pred))
    return {'Model': model_name, 'MAE': mae, 'RMSE': rmse}

# --- Gather Results ---
results = [
    evaluate(test['Citations'], lr_preds, 'Linear Regression'),
    evaluate(test['Citations'], holt_preds, 'Holt Linear Trend'),
    evaluate(test['Citations'], arima_preds, 'ARIMA'),
    evaluate(test['Citations'], rf_preds, 'Random Forest'),
    evaluate(test['Citations'], prophet_preds, 'Prophet')
]

results_df = pd.DataFrame(results).sort_values('RMSE')
print("\nðŸ“Š Accuracy Comparison:")
print(results_df)

# --- Forecast 2026â€“2030 using Best Model ---
best_model = results_df.iloc[0]['Model']
future_years = np.arange(2026, 2031)

if best_model == 'ARIMA':
    final_model = ARIMA(citations_df['Citations'], order=(2,1,2)).fit()
    future_preds = final_model.forecast(len(future_years))

elif best_model == 'Holt Linear Trend':
    final_model = ExponentialSmoothing(citations_df['Citations'], trend='add', seasonal=None).fit()
    future_preds = final_model.forecast(len(future_years))

elif best_model == 'Linear Regression':
    final_model = LinearRegression()
    final_model.fit(citations_df[['Year']], citations_df['Citations'])
    future_preds = final_model.predict(future_years.reshape(-1, 1))

elif best_model == 'Random Forest':
    final_model = RandomForestRegressor(random_state=42)
    final_model.fit(citations_df[['Year']], citations_df['Citations'])
    future_preds = final_model.predict(future_years.reshape(-1, 1))

elif best_model == 'Prophet':
    full_prophet_df = citations_df[['Year', 'Citations']].rename(columns={'Year': 'ds', 'Citations': 'y'})
    full_prophet_df['ds'] = pd.to_datetime(full_prophet_df['ds'], format='%Y')

    final_prophet = Prophet()
    final_prophet.fit(full_prophet_df)

    future_df = pd.DataFrame({'ds': pd.to_datetime(future_years, format='%Y')})
    forecast_future = final_prophet.predict(future_df)
    future_preds = forecast_future['yhat'].values

# --- Plot Full Forecast ---
plt.figure(figsize=(14, 7))
plt.plot(citations_df['Year'], citations_df['Citations'], label='Historical Citations', marker='o')
plt.plot(test['Year'], test['Citations'], label='Actual Test (2021â€“2025)', marker='o')

# Forecasted values for 2026â€“2030
plt.scatter(future_years, future_preds, color='red', label=f'{best_model} Prediction (2026â€“2030)', zorder=5)

plt.title(f"Citations Forecast with {best_model} (2026â€“2030)")
plt.xlabel("Year")
plt.ylabel("Citations")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# --- Print Forecast Table ---
forecast_table = pd.DataFrame({'Year': future_years, 'Predicted Citations': future_preds})
print(f"\nðŸ“ˆ {best_model} Forecast 2026â€“2030:")
print(forecast_table)
```

```python
import pandas as pd
import numpy as np
from prophet import Prophet
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt

# Load data
file_path = "author_details_output.xlsx"  # Adjust if needed
df = pd.read_excel(file_path, sheet_name='Citations Per Year')

# Prepare time series data
df = df.rename(columns={'Year': 'ds', 'Citations': 'y'})
df['ds'] = pd.to_datetime(df['ds'], format='%Y')
df.set_index('ds', inplace=True)

# Train-test split: Train till 2022, test 2023â€“2025
train = df[df.index.year <= 2022]
test = df[df.index.year > 2022]

# ======= 1. Prophet Model =======
prophet_df = train.reset_index().rename(columns={'ds': 'ds', 'y': 'y'})
model_prophet = Prophet()
model_prophet.fit(prophet_df)

future = model_prophet.make_future_dataframe(periods=len(test), freq='Y')
forecast = model_prophet.predict(future)

# Extract forecast only for test years
forecast['year'] = forecast['ds'].dt.year
prophet_pred = forecast.set_index('year').loc[test.index.year, 'yhat']

# Ensure test actual also indexed by year
test_actual = test.copy()
test_actual.index = test_actual.index.year

# ======= 2. ARIMA Model =======
arima_model = ARIMA(train['y'], order=(1, 1, 1))  # You can tune (p,d,q)
arima_result = arima_model.fit()
arima_pred = arima_result.forecast(steps=len(test))
arima_pred.index = test.index.year

# ======= 3. Holt-Winters Exponential Smoothing =======
hw_model = ExponentialSmoothing(train['y'], trend='add', seasonal=None)
hw_result = hw_model.fit()
hw_pred = hw_result.forecast(steps=len(test))
hw_pred.index = test.index.year

# ======= Evaluation Function =======
def evaluate(y_true, y_pred, model_name):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = mean_squared_error(y_true, y_pred, squared=False)
    print(f"{model_name} â†’ MAE: {mae:.2f}, RMSE: {rmse:.2f}")
    return mae, rmse

# ======= Accuracy Comparison =======
print("\nðŸ“Š Accuracy Comparison on 2023â€“2025:\n")
evaluate(test_actual['y'], prophet_pred, "Prophet")
evaluate(test_actual['y'], arima_pred, "ARIMA")
evaluate(test_actual['y'], hw_pred, "Holt-Winters")

# ======= Plotting =======
plt.figure(figsize=(14, 6))
plt.plot(df.index.year, df['y'], label='Actual', color='black', marker='o')
plt.plot(prophet_pred.index, prophet_pred, label='Prophet', color='skyblue', marker='o')
plt.plot(arima_pred.index, arima_pred, label='ARIMA', color='green', marker='x')
plt.plot(hw_pred.index, hw_pred, label='Holt-Winters', color='orange', marker='s')
plt.title("Citations Forecast Comparison (2023â€“2025)")
plt.xlabel("Year")
plt.ylabel("Number of Citations")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
