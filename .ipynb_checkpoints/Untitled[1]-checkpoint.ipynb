```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv("heart disease dataset.csv")

# Calculate average citations per year
avg_citations_per_year = df.groupby('Year')['Citations'].mean().reset_index()
avg_citations_per_year.columns = ['Year', 'Average Citations']

# Plotting
plt.figure(figsize=(10, 5))
sns.lineplot(data=avg_citations_per_year, x='Year', y='Average Citations', marker='o')
plt.title('Average Citations Per Article Per Year')
plt.xlabel('Year')
plt.ylabel('Average Citations')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx
from collections import Counter

# Load your dataset
df = pd.read_csv("heart disease dataset.csv")

# --- 1. Most Prolific Authors by Number of Publications ---
author_publication_count = Counter()
for authors in df['Authors'].dropna():
    for author in authors.split(','):
        author_publication_count[author.strip()] += 1

top_authors_by_pub = pd.DataFrame(author_publication_count.items(), columns=['Author', 'Publication Count'])
top_authors_by_pub = top_authors_by_pub.sort_values(by='Publication Count', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(data=top_authors_by_pub, x='Publication Count', y='Author')
plt.title('Top 10 Authors by Number of Publications')
plt.show()

# --- 2. Co-authorship Network Graph ---
G = nx.Graph()
for authors in df['Authors'].dropna():
    author_list = [a.strip() for a in authors.split(',')]
    for i in range(len(author_list)):
        for j in range(i + 1, len(author_list)):
            a1, a2 = author_list[i], author_list[j]
            if G.has_edge(a1, a2):
                G[a1][a2]['weight'] += 1
            else:
                G.add_edge(a1, a2, weight=1)

# Draw a subgraph of top 30 authors by degree
top_nodes = sorted(G.degree, key=lambda x: x[1], reverse=True)[:30]
subgraph = G.subgraph([node for node, _ in top_nodes])

plt.figure(figsize=(12, 12))
pos = nx.spring_layout(subgraph, k=0.5)
nx.draw(subgraph, pos, with_labels=True, node_size=500, font_size=8,
        node_color='skyblue', edge_color='gray')
plt.title('Co-authorship Network (Top 30 Authors)')
plt.show()

# --- 3. Citation Distribution Histogram ---
plt.figure(figsize=(10, 5))
sns.histplot(df['Citations'], bins=30, kde=True)
plt.title('Citation Distribution of Articles')
plt.xlabel('Citations')
plt.ylabel('Number of Articles')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'heart disease dataset.csv'
df = pd.read_csv(file_path)

# Preview the first few rows
print(df.head())

# -------------------------------
# 1. Compare average citations per publisher
# -------------------------------

# Group by publisher and calculate average citations
avg_citations = df.groupby('Publisher')['Citations'].mean().reset_index()

# Sort by average citations for better readability
avg_citations = avg_citations.sort_values(by='Citations', ascending=False)

# Plot average citations per publisher
plt.figure(figsize=(12, 6))
sns.barplot(data=avg_citations, x='Citations', y='Publisher', palette='viridis')
plt.title('Average Citations per Publisher')
plt.xlabel('Average Citations')
plt.ylabel('Publisher')
plt.tight_layout()
plt.show()

import numpy as np

# Create random data for this plot
random_data = np.random.rand(len(df))
df['RandomCitations'] = random_data

# Group by Publisher and calculate the sum of random values (it doesn't have any meaningful interpretation)
random_citations = df.groupby('Publisher')['RandomCitations'].sum().reset_index()

# Plot random citations per publisher
plt.figure(figsize=(12, 6))
sns.barplot(data=random_citations, x='RandomCitations', y='Publisher', palette='magma')
plt.title('Unclear Random Citations per Publisher')
plt.xlabel('Random Citations')
plt.ylabel('Publisher')
plt.tight_layout()
plt.show()

# -------------------------------
# 2. Trends in publications per publisher over time
# -------------------------------

# Group by year and publisher, count number of publications
pub_trends = df.groupby(['Year', 'Publisher']).size().reset_index(name='Publication_Count')

# Plot publication trends over time
plt.figure(figsize=(14, 7))
sns.lineplot(data=pub_trends, x='Year', y='Publication_Count', hue='Publisher', marker='o')
plt.title('Publication Trends per Publisher Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Publications')
plt.legend(title='Publisher', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.arima.model import ARIMA
from prophet import Prophet

# Load the dataset
file_path = 'heart disease dataset.csv'
df = pd.read_csv(file_path)

# Preview the first few rows
# print(df.head())

# -------------------------------
# Total citations per year
# -------------------------------

# Group total citations by year
citations_per_year = df.groupby('Year')['Citations'].sum().reset_index()

# Plot total citations per year
plt.figure(figsize=(12, 6))
sns.lineplot(data=citations_per_year, x='Year', y='Citations', marker='o')
plt.title('Total Citations per Year')
plt.xlabel('Year')
plt.ylabel('Total Citations')
plt.grid(True)
plt.tight_layout()
plt.show()

# -------------------------------
# ARIMA Forecasting
# -------------------------------

# Set 'Year' as index for time series
citations_ts = citations_per_year.set_index('Year')['Citations']

# Fit ARIMA model (order can be tuned after testing ACF/PACF)
model_arima = ARIMA(citations_ts, order=(1, 1, 1))
result_arima = model_arima.fit()

# Forecast next 5 years
forecast_arima = result_arima.get_forecast(steps=5)
forecast_index = range(citations_ts.index.max()+1, citations_ts.index.max()+6)

# Plot original + forecast
# plt.figure(figsize=(12, 6))
# plt.plot(citations_ts, label='Observed')
# plt.plot(forecast_index, forecast_arima.predicted_mean, label='ARIMA Forecast', linestyle='--')
# plt.fill_between(forecast_index, 
#                  forecast_arima.conf_int().iloc[:, 0], 
#                  forecast_arima.conf_int().iloc[:, 1], color='lightgrey', alpha=0.5)
# plt.title('Citations Forecast with ARIMA')
# plt.xlabel('Year')
# plt.ylabel('Total Citations')
# plt.legend()
# plt.grid(True)
# plt.tight_layout()
# plt.show()

# -------------------------------
# Prophet Forecasting
# -------------------------------

# Prepare data for Prophet
prophet_df = citations_per_year.rename(columns={'Year': 'ds', 'Citations': 'y'})

# Initialize and fit model
model_prophet = Prophet(yearly_seasonality=True)
model_prophet.fit(prophet_df)

# Create future dataframe for next 5 years
future = model_prophet.make_future_dataframe(periods=5, freq='Y')

# Predict
forecast_prophet = model_prophet.predict(future)

# Plot forecast
fig = model_prophet.plot(forecast_prophet)
plt.title('Citations Forecast with Prophet')
plt.xlabel('Year')
plt.ylabel('Total Citations')
plt.tight_layout()
plt.show()
```
